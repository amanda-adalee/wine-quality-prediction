```python
import pandas
import numpy
from matplotlib import pyplot
import seaborn
import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import ipywidgets
import warnings
warnings.filterwarnings('ignore')
```

# READ DATASET

```python
wine_data = pandas.read_csv("winequality-red.csv")
wine_data.head(10)
```

# DATA CLEAN UP

```python
wine_data.shape # THERE ARE 1599 WINES(ROWS) AND 12 ATTRIBUTES(COLUMNS)
```

```python
wine_data.isnull().sum() # CHECKS TO SEE THE AMOUNT OF CELLS THAT ARE NULL (NO NULLS)
```

```python
wine_data.duplicated().sum()#CHECKS TO SEE HOW MANY DUPLICATE WINES THERE ARE (240)
```

```python
wine_data=wine_data.drop_duplicates(keep='first') #DELETES DUPLICATES
```

```python
wine_data.shape
```

```python
# DUPLICATES HAVE BEEN DELETED. THERE ARE NO NULL CELLS. THERE ARE 1359 WINES(ROWS) AND 12 ATTRIBUTES(COLUMNS)
```

# REMOVE OUTLIERS

```python
q1=wine_data.quantile(0.25)
q2=wine_data.quantile(0.75)
IQR=q2-q1
```

```python
wine_data=wine_data[~((wine_data<(q1-1.5*IQR))|(wine_data>(q2+1.5*IQR))).any(axis=1)]
```

# STANDARDIZE QUALITY RATING

```python
# MAP RATING TO BAD(0), MID(1), GOOD(2) 
bins = [0, 6.5, 10] # THIS MEANS 3-6 ARE POOR, 7-9 ARE HIGH
labels = [0, 1]
wine_data['quality'] = pandas.cut(wine_data['quality'], bins=bins, labels=labels)
```

```python
wine_data.head(10)
```

# TRAINING AND TESTING DATA SPLIT

```python
x = wine_data[wine_data.columns[:-1]]
y = wine_data['quality']
sc = StandardScaler()
x = sc.fit_transform(x)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2, random_state=42)
```

```python
for data in [y_train, y_test]:
    print(data.describe())
```

# CLASSIFICATION

**K NEAREST NEIGHBOR**

```python
nn3 = KNeighborsClassifier(n_neighbors = 3)
nn3.fit(x_train, y_train)
pred = nn3.predict(x_test)
print(classification_report(y_test, pred))
print("KNeighborsClassifier model has an accuracy of {}".format(round(accuracy_score(y_test, pred),2)))
```

**C-SUPPORT VECTOR**

```python
svc = SVC()
svc.fit(x_train, y_train)
pred = svc.predict(x_test)
print(classification_report(y_test, pred))
print("C-Support Vector Classification model has an accuracy of {}".format(round(accuracy_score(y_test, pred),2)))
```
